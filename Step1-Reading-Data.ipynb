```python
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd
import zipfile
```

# Step 1: Reading the data

The database used here is the [Flowers-299](https://www.kaggle.com/datasets/bogdancretu/flower299/) kaggle dataset. \
To run this notebook download the dataset from there and save it as 'images.zip'

```python
data = []
with zipfile.ZipFile('images.zip') as f:
    for name in f.namelist():
        _, class_name, filename = name.split('/')
        img = Image.open(f.open(name))
        data.append({"Class": class_name, "Filename": filename, "Img": img})
df = pd.DataFrame.from_dict(data)
df
```

```python
# Print out a few samples, to check that the images are read correctly.
fig, ax = plt.subplots(3,3, figsize=(15, 15))
r, c = 0, 0
for _, row in df.sample(n=9).iterrows():
    ax[r, c].imshow(row['Img'])
    ax[r, c].title.set_text(row['Class'])
    
    r += 1
    if r == 3:
        r = 0
        c += 1
```
